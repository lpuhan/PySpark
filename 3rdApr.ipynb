```python
gVar1 ="Testing Global variable"
gVar2 ="Testing Global variable"
def testDeclaration():
    try:
        # global gVar
        lVar ="Inside local var testDeclaration"
        print(lVar)
        gVar1+=lVar 
        print(gVar1)
    except UnboundLocalError:
        print('UnboundLocalError')
    finally:
        print('in Finally block')
        global gVar2 
        l1Var ="Inside local var testDeclaration"
        print(lVar)
        gVar2+=lVar 
        print(gVar2)
    


testDeclaration()
```
